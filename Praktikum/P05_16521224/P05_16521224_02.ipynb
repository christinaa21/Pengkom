Deskripsi: Diberikan data pengguna kartu kredit suatu bank. Menuliskan:

            1. Jumlah pengguna tidak aktif (Attrited Customer) yang berusia di atas 50 tahun.

            2. ID pengguna dengan nominal limit kartu kredit terbesar.

            3. Jumlah pengguna dengan tipe kartu Blue dan lama penggunaan kartu kredit dalam bulan di atas 40.

            4. Banyaknya pengguna dengan total nominal yang belum dibayar pengguna kartu kredit sama dengan 0 untuk masing-masing kategori penghaslan.

            5. 5 ID pengguna aktif (Existing Customer) dengan banyak bulan kartu kredit tidak aktif dalam 12 bulan terakhir terbanyak.
           

**Import Data**

```python
import pandas as pd
df = pd.read_csv('P05-01.csv')
```

**1. Jumlah pengguna tidak aktif (Attrited Customer) yang berusia di atas 50 tahun.**

```python
attrited = df.loc[(df['Attrition_Flag'] == 'Attrited Customer') & (df['Customer_Age'] > 50)]
print('Jumlah pengguna tidak aktif yang berusia di atas 50 tahun =', len(attrited))
attrited
```

**2. ID pengguna dengan nominal limit kartu kredit terbesar.**

```python
# Cara 1 (Menampilkan semua kolom)
df.loc[df['Credit_Limit'] == df['Credit_Limit'].max()]
```

```python
# Cara 2 (Menampilkan hanya kolom CLIENTNUM dan Credit_Limit saja)
df.loc[df['Credit_Limit'] == df['Credit_Limit'].max(), ['CLIENTNUM', 'Credit_Limit']]
```

```python
# Cara 3 (Menuliskan list dari ID pengguna)
ID = df.loc[df['Credit_Limit'] == df['Credit_Limit'].max(), ['CLIENTNUM']]
list_ID = ID['CLIENTNUM'].to_list()
print('List ID pengguna dengan nominal limit kartu kredit terbesar (', df['Credit_Limit'].max(), '):', list_ID)
```

**3. Jumlah pengguna dengan tipe kartu Blue dan lama penggunaan kartu kredit dalam bulan di atas 40.**

```python
blue_40 = df.loc[(df['Card_Category'] == 'Blue') & (df['Months_on_book'] > 40)]
print('Jumlah pengguna dengan tipe kartu Blue dan lama penggunaan kartu kredit dalam bulan di atas 40 =', len(blue_40), 'pengguna')
blue_40
```

**4. Banyaknya pengguna dengan total nominal yang belum dibayar pengguna kartu kredit sama dengan 0 untuk masing-masing kategori penghaslan.**

```python
df.loc[df['Total_Revolving_Bal'] == 0]['Income_Category'].value_counts()
```

**5. 5 ID pengguna aktif (Existing Customer) dengan banyak bulan kartu kredit tidak aktif dalam 12 bulan terakhir terbanyak.**

```python
# Cara 1 (Menampilkan semua kolom)
df.loc[(df['Attrition_Flag'] == 'Existing Customer') & (df['Months_Inactive_12_mon'] == df['Months_Inactive_12_mon'].max())][:5]
```

```python
# Cara 2 (Hanya menampilkan kolom CLIENTNUM, Attrition_Flag, Months_Inactive_12_mon)
df.loc[(df['Attrition_Flag'] == 'Existing Customer') & (df['Months_Inactive_12_mon'] == df['Months_Inactive_12_mon'].max()), ['CLIENTNUM', 'Attrition_Flag', 'Months_Inactive_12_mon']][:5]
```

```python
# Cara 3 (Menuliskan list dari ID pengguna)
ID1 = df.loc[(df['Attrition_Flag'] == 'Existing Customer') & (df['Months_Inactive_12_mon'] == df['Months_Inactive_12_mon'].max()), ['CLIENTNUM']][:5]
list_ID1 = ID1['CLIENTNUM'].to_list()
print('List ID pengguna dengan nominal limit kartu kredit terbesar (', df['Months_Inactive_12_mon'].max(), '):', list_ID1)
```
