**Import Data**

```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv('H05.csv')
```

**1. Histogram Distribusi Nilai Matematika**

```python
df[["nilai_mat"]].plot(kind="hist",bins=[20,30,40,50,60,70,80,90,100],rwidth=0.9,title='Distribusi Nilai Matematika')
plt.show()
```

**2. Diagram Batang Horizontal Banyaknya Mahasiswa Masing-Masing Fakultas**

```python
df1 = df['fakultas'].value_counts()
df1.plot(kind="barh", title="Banyaknya Mahasiswa Masing-Masing Fakultas")
plt.show()
```

**3. Diagram Pie Banyaknya Mahasiswa Masing-Masing Fakultas**

```python
df1 = df['fakultas'].value_counts()
df1.plot(kind="pie", title="Banyaknya Mahasiswa Masing-Masing Fakultas")
plt.show()
```

**4. Penjelasan diagram yang lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak**

```python
print("Berdasarkan diagram pie dan batang horizontal, fakultas dengan mahasiswa terbanyak adalah FTI.\n" +
      "Menurut saya, grafik batang horizontal lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak, \n" +
      "karena lebih mudah dilihat manakah batang yang paling panjang sebagai simbol dari fakultas dengan mahasiswa terbanyak.\n" +
      "Sedangkan, jika menggunakan pie chart, kita tidak bisa terlalu memastikan manakah sudut yang terbesar kecuali jika terdapat data persentasenya.")
```

**5. Scatter Plot dengan Nilai Kimia sebagai x dan Fisika sebagai y**

```python
df.plot(kind="scatter", x='nilai_kim', y='nilai_fis', title="Hubungan Nilai Kimia dan Nilai Fisika")
plt.show()
```
