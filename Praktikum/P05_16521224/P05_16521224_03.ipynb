Deskripsi: Diberikan data harga rumah yang disewakan di Airbnb di Singapura. Menuliskan:

            1. Histogram distribusi banyaknya review rumah yang disewakan di wilayah (neighbourhood group) West Region dengan bins 20.

            2. Pie chart perbandingan banyaknya masing-masing neighbourhood dengan harga rumah per malam di atas 300.

            3. Scatter plot dengan long sebagai x dan lat sebagai y dari rumah yang ada di neighbourhood Kallang.

            4. Bubble plot dengan long sebagai x, lat sebagai y, dan banyaknya review sebagai ukuran dari rumah dengan banyak hari ketersediaan rumah dalam satu tahun di atas 100.

            5. Stacked bar chart dari rumah yang disewakan oleh Richards dengan neighbourhood sebagai x dan jenis penyediaan ruang sebagai stacked y.
           

**Import Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('P05-03.csv')
```

**1. Histogram distribusi banyaknya review rumah yang disewakan di wilayah (neighbourhood group) West Region dengan bins 20.**

```python
review_west_region = df.loc[df['neighbourhood_group'] == 'West Region']
review_west_region[["number_of_reviews"]].plot(kind="hist",bins=[0,20,40,60,80,100],rwidth=0.9,title='Distribusi Banyaknya Review Rumah yang Disewakan di Wilayah West Region')
plt.show()
```

**2. Pie chart perbandingan banyaknya masing-masing neighbourhood dengan harga rumah per malam di atas 300.**

```python
df.loc[df['price'] > 300]['neighbourhood'].value_counts().plot(kind='pie', title='Perbandingan Banyaknya Masing-Masing Neighbourhood dengan Harga Rumah per Malam di Atas 300')
plt.show()
```

**3. Scatter plot dengan long sebagai x dan lat sebagai y dari rumah yang ada di neighbourhood Kallang.**

```python
df.loc[df['neighbourhood'] == 'Kallang'].plot(kind = 'scatter', x='longitude', y='latitude', title='Hubungan antara Longitude dan Latitude Rumah yang Ada di Neighbourhood Kallang')
plt.show()
```

**4. Bubble plot dengan long sebagai x, lat sebagai y, dan banyaknya review sebagai ukuran dari bubble dengan banyak hari ketersediaan rumah dalam satu tahun di atas 100.**

```python
df.loc[df['availability_365'] > 100].plot(kind='scatter', x='longitude', y='latitude', sizes=df['number_of_reviews'], title='Hubungan antara Longitude, Latitude, dan Banyaknya Review dengan Banyak Hari Ketersediaan Rumah dalam Satu Tahun di Atas 100')
plt.show()
```

**5. Stacked bar chart dari rumah yang disewakan oleh Richards dengan neighbourhood sebagai x dan jenis penyediaan ruang sebagai stacked y.**

```python
df.loc[df['host_name'] == 'Richards']['neighbourhood'].value_counts().plot(kind='bar', x='neighbourhood', y='room_type', title='Rumah yang Disewakan oleh Richards', stacked='True')
plt.show()
```
